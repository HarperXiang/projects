# Web Scrapping from Wikipedia on S&P 500 Companies List

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(url)
len(tables)
```

```python
for i in range(len(tables)):
    df = tables[i]
    print(f"Table [{i}]#################################")
    print(df.head())
```

```python
df500 = tables[0].iloc[1:]
df500.columns = ['Symbol', 'Security', 'SEC filings', 'GICS Sector',
                 'GICS Sub Industry', 'Headquarters Location', 'Date first added',
                 'CIK', 'Founded']
# df500.set_index('Symbol', inplace=True)
df500.head()
```

```python
len(df500)
```

```python
datapath = "data/"
filename = "S&P500_list.csv"
df500.to_csv(datapath+filename)
```
