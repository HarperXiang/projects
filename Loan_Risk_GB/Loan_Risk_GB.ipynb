# LGD Modeling
LGD: ONLY keep charged off: loss= loan amount – total payment ,
Regression to predict the loss, mse, mae, 


```python
dictonary=pd.read_csv('data/Target_Leakage.csv', encoding= 'unicode_escape')
features_delete=list(dictonary[dictonary['Target Leakage?']!='No']['LoanStatNew'].values)

features_delete.remove('loan_status')
features_delete.remove('total_pymnt')
features_delete.remove('funded_amnt')
drop_features=[i for i in features_delete if i in list(LC1.columns)]
data_selected=LC1.drop(columns=drop_features)

# Only select data charged off

data_selected=data_selected[data_selected['loan_status']=='Charged Off']
data_selected['LGD'] = data_selected['total_pymnt']-data_selected['funded_amnt']

missing_col=data_selected.isna().sum()
missing_features=list(missing_col[missing_col!=0].index)
data=data_selected.drop(columns=missing_features)
print(data.shape)

data_LGD=data.drop(columns=['loan_status','url','total_pymnt','funded_amnt'])

# Get dummies with LCdata2
L2_features=[i for i in list(LC2.columns) if i in list(data_LGD.columns)]
L2_features

L2_LGD=LC2[L2_features]
print(L2_LGD.shape)
L2_LGD=L2_LGD.dropna()
print(L2_LGD.shape)
data_merged_LGD=pd.concat([data_LGD,L2_LGD],axis=0)
data_merged_LGD=pd.get_dummies(data_merged_LGD)
print(data_merged_LGD.shape)
data_LGD=data_merged_LGD.iloc[:data_LGD.shape[0],:]
L2_LGD=data_merged_LGD.iloc[data_LGD.shape[0]:,:]
print(data_LGD.shape,L2_LGD.shape)

L2_LGD=L2_LGD.drop(columns=['LGD'])
print(L2_LGD.shape)

```

# Task4. 
- Evaluation on the model performances
- Given: Costs as extra \\$1,000 in processing. Every time we reject someone that would have been a good loan, we lose \\$2,000 in additional revenue  

```python
def custom_score(y_true, y_pred):
    y_true=np.array(y_true)
    y_pred=np.array(y_pred)
    if (y_true.size != y_pred.size):
        print('prediction and true values No match')
        
    loss=0
    for i in range(len(y_true)):
        if ((y_true[i]==1) and (y_pred[i]==0)):  # incorrectly make a loan that’s charged off or defaulted
            loss += 1000
        elif ((y_true[i]==0) and (y_pred[i]==1)): # reject someone that would have been a good loan
            loss += 2000
        else:
            loss += 0
    return loss
```

```python
model2_xgb=XGBClassifier(learning_rate=0.02, n_estimators=400, random_state=42,max_depth=3,
              scale_pos_weight=4.874708574389086).fit(X_train, y_train, eval_metric=custom_score)

xgb2_pred=model2_xgb.predict(X_test)
print("confusion matrix for for Gradient Boosting :")
print(confusion_matrix(y_test,xgb2_pred))
print("classification report for for Gradient Boosting : ")
print(classification_report(y_test,xgb2_pred))
```

# Task6. 
- Portfolio capital risk estimation
- Given all the loans made in Task2.

```python
#Since our model is built based on the LC data, we are assuming it can be served as a proxy of how well our future portfolio will run,
#With this assumption, we will calculate the gain/loss for the LC data and then run Monte Carlo simulation off it
#in order to calculate the VaR of 95%.

data_q6=pd.read_csv('data/LCdata.csv')
loan_status_excluded = ["Current",
                        "Does not meet the credit policy. Status:Fully Paid",
                        "Does not meet the credit policy. Status:Charged Off",
                        "In Grace Period"]
data_q6=data_q6[~np.isin(data_q6["loan_status"].values, loan_status_excluded)]
print(data_q6.shape)


data_q6['Default'] = data_q6['loan_status'].apply(lambda x : 1 if x == 'Charged Off' else 0)

data_q6['term_num'] = data_q6.term.str.extract('(\d+)').astype(int)

data_q6.loc[data_q6['Default'] == 0, 'Loan_Profit_annualized'] = (data_q6['total_rec_int'])/ data_q6['term_num'] *12
data_q6.loc[data_q6['Default'] == 1, 'Loan_Profit_annualized'] = (data_q6['total_pymnt'] - data_q6['funded_amnt'])
```

```python
portfolio_profit_for_sim = data_q6['Loan_Profit_annualized'].values
resamples = np.random.choice(portfolio_profit_for_sim, size=(1000, portfolio_profit_for_sim.shape[0]), replace=True)
resamples = resamples.T
sim_loan_profit = resamples.sum(axis=0)
```

```python
Var_95 = np.percentile(sim_loan_profit, 5)
print('VaR for 95th percentile: ',  Var_95)
```

# Task7. 
- If there’s a downturn in the economy, how this might affect both the individual performance of our loans and our portfolio overall
- How to account for something like this in our models and how this changes the performance of our portfolio and our capital requirements in at least a couple of different scenarios

<span style='color:Green'>
In a bad economic scenario, individual Default Probabilities are supposed to increase, because people's incomes may decline and thus their payment capabilities may be weakened. The effects can be captured by our Default Probability model's variables. For example, if a person relies more on credit cards to pay bills when his income declines, that will result in a higher credit balance rate, which will push up his Default Probabilities reflected in our model.
    
However, the existing model may become inaccurate for our current portfolio of 50K loans when the overall scenario significantly changes. That is saying, the value distributions of those variables have changed significantly, and as a result, collectively push the overall Default ratio of a loan portfolio going upwards. To estimate this kind of effects, we need the process:
    
- First, we need to know the how much the overall increase of Default ratio. And this can be done by using a separate model that introduces macroeconomic data as predictors. Assuming now we already get the prediction that the Default ratio will increase 1%, 5% or 20%.
- Then, we can estimate the effects on our current loan portfolio. Instead of using the predicted Default Probabilities and LGDs by the existing models, we use the Monte Carlo simulation with this updated Default ratio. LGDs are simulated by resampling from the historical portfolio and then added to the expected gain of the current portfolio. The estimation of our current portfolio's gain or loss will correspond to the updated Default ratio.
- We need to simulate a batch of these homogeneous portfolios and then observe the distribution of their expected gain or loss. This approach will help us understand the range of the effects of changes and capital requirements in facing such changes.
    
The next we will demonstrate the said process to estimate the overall effects on our current portfolio.
<span>

### . Preparing Data

```python
usecols = ["loan_status", "funded_amnt", "total_pymnt"]
hist_portfolio = pd.read_csv('data/LCdata.csv', usecols=usecols)
print(hist_portfolio.shape)

usecols = ["funded_amnt", "int_rate", "term"]
curr_portfolio = pd.read_csv('data/LCdata2.csv', usecols=usecols)
print(curr_portfolio.shape)
```

```python
# Historical Portfolio

# "Default"
loan_status_excluded = ["Current",
                        "Does not meet the credit policy. Status:Fully Paid",
                        "Does not meet the credit policy. Status:Charged Off",
                        "In Grace Period"]
hist_portfolio = hist_portfolio[~np.isin(hist_portfolio["loan_status"].values, loan_status_excluded)]
hist_portfolio['Default'] = hist_portfolio['loan_status'].apply(lambda x : 1 if x == 'Charged Off' else 0)

# "LGD", "LGD_ratio"
hist_portfolio["LGD"] = hist_portfolio["total_pymnt"] - hist_portfolio["funded_amnt"]
hist_portfolio["LGD_ratio"] = hist_portfolio["LGD"] / hist_portfolio["funded_amnt"]

print(hist_portfolio.shape)
```

```python
# Current Portfolio

# "Term"
curr_portfolio = curr_portfolio.dropna()
curr_portfolio["term"] = curr_portfolio["term"].str.extract('(\d+)').astype(int)
curr_portfolio["term"] = curr_portfolio["term"]/12

# "Gain"
curr_portfolio["Gain"] = curr_portfolio["funded_amnt"] * \
                         ((1.0+curr_portfolio["int_rate"]/100)**curr_portfolio["term"]) \
                         - curr_portfolio["funded_amnt"]
curr_portfolio["Gain_annual"] = curr_portfolio["Gain"] / curr_portfolio["term"] 

print(curr_portfolio.shape)
```

```python
hist_default_ratio = round(len(hist_portfolio[hist_portfolio["Default"]==1]) / len(hist_portfolio), 5)
print(f'Historical Default Ratio: [{100*hist_default_ratio}%]')

hist_data = hist_portfolio[hist_portfolio["Default"]==1]["LGD_ratio"]
display(hist_data.head())
print(hist_data.shape)

curr_data = curr_portfolio[["funded_amnt", "Gain_annual"]]
display(curr_data.head())
print(curr_data.shape)
```

### . Monte Carlo Simulation

```python
# FUNCTION
# Monte Carlo Simulation of a Portfolio of Loans
import time

def MC_portfolio (curr_data, hist_data, ratio_default):

#     timer_start = time.time()

    # Get statistics of population
    num_loans      = curr_data.shape[0]
    num_lgd        = hist_data.shape[0]

    # TO save running time, the Default probabilities for a Portfolio is simulated for once
    # Instead of simulating a single Default probability for many times
    # The effects of both approaches are the same
    default_probabilities = np.random.rand(num_loans)

    # Portfolio statistics
    loan_total  = 0.0
    gain_total  = 0.0
    gain_annual = 0.0

    # Simulating Gain of Current Portfolio
    for idx in range(num_loans):
        
        # Simulating a LGD with a historical LGD_ratio sample and the current Loan_amount
        if (default_probabilities[idx] <= ratio_default):
            sample_idx = np.random.choice(num_lgd, size=1)[0]
            gain_annual += hist_data[sample_idx] * curr_data[idx, 0]
            
        # Simulating the Gain of the current Non-Default data
        else:
            gain_annual += curr_data[idx, 1]
            
        loan_total  += curr_data[idx, 0]
        
    # Calculating Portfolio's Loss (or Gain)
    portfolio_stat = {"loan_total"  : loan_total, 
                      "gain_annual" : gain_annual}

#     timer_end = time.time()
#     print(f'Time for {num_loans} samples: {round((timer_end-timer_start)/60, 2)} minutes')
    return (portfolio_stat)
```

```python
# FUNCTION
# N times of MC Simulation

def N_Simulations (MC_num, curr_data, hist_data, ratio_default):

    timer_start = time.time()

    # Run N times MC_portfolio
    portfolio_stats = []
    for i in range(MC_num):
        portfolio_stat = MC_portfolio (curr_data, hist_data, ratio_default)
        portfolio_stats.append(portfolio_stat)
        
    # Return Statistics
    loan_total  = []
    gain_annual = []
    for i in range(MC_num):
        loan_total.append(portfolio_stats[i]["loan_total"])
        gain_annual.append(portfolio_stats[i]["gain_annual"])

    df_portfolio_stats = pd.DataFrame({"loan_total"  : loan_total, 
                                       "gain_annual" : gain_annual})
    df_portfolio_stats["gain_ratio_annual"] = df_portfolio_stats["gain_annual"] / df_portfolio_stats["loan_total"]
    df_portfolio_stats = df_portfolio_stats.sort_values(["gain_ratio_annual"], ascending=True)

    timer_end = time.time()
    print(f'Time for simulating {MC_num} Portfolios: {round((timer_end-timer_start)/60, 2)} minutes')
    return (df_portfolio_stats)
```

```python
# FUNCTION
# Print Statistics of N Simulated Portfolios

def print_portfolios (df_portfolio_stats, quantiles=[0.05, 0.5, 0.95]):

    for q in quantiles:
        print(f'Estimatd Annual Gain at quantile({q}): ${round(df_portfolio_stats["gain_annual"].quantile(q), 2)}')

    plt.figure(figsize=(15,5))
#     plt.subplot(2,2,1)
#     sns.distplot(df_portfolio_stats["loan_total"])
#     plt.xlabel("loan_total")
#     plt.ylabel("Frequency")
   
    plt.subplot(1,2,1)
    sns.distplot(df_portfolio_stats["gain_annual"])
    plt.xlabel("gain_annual")
    plt.ylabel("Frequency")

    plt.subplot(1,2,2)
    sns.distplot(df_portfolio_stats["gain_ratio_annual"])
    plt.xlabel("gain_ratio_annual")
    plt.ylabel("Frequency")
    plt.show()
    
    return
```
