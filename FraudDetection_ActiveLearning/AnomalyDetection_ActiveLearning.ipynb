## Project Overview

This project is to detect anomalies from un-labeled data using Active Learning techniques. Detecting anomalies is binary classification, but as the target variable is assumed unknown, the detecting process has to be Unsupervised Learning. The goal to be achieved is labeling as many anomalies with as less human labeling effort. In this project, the given conditions are:
- Limited budget of human labeling
- Imbalanced data (the degree of imbalance is assumed unknown)

The solution is constructed upon three pillars:
- Unsupervised Learning

With the large amount of un-labeled data, Neural Network is available for Unsupervised Learning. AutoEncoder models are adopted to detect anomalies by reconstructing X and checking the reconstruction errors, mean squared error (MSE) in this case.

- Supervised Learning

When human label involved, supervised learning is available. But with the tiny volume of labeled data, only GLM or Tree models are possible. XGBClassifier and Logistic models are both used for specific purposes in this project. Given the data amount, overfitting could be an issue. So it is crucial to know where the limits are and take advantage of Supervised Learning in innovative ways.

- Integrating the two parts

This is the pivot of the solution. Neither Unsupervised nor Supervised Learning can solely resolve the challenges. Appropriately combining the two techniques may successfully implement Active Learning to complete the task.

### Data Source Statement
The dataset is the Credit Card Fraud data from Kaggle (https://www.kaggle.com/mlg-ulb/creditcardfraud/data#).
The dataset contains transactions made by credit cards in September 2013 by european cardholders.
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

It contains only numerical input variables which are the result of a PCA transformation. Unfortunately, due to confidentiality issues, we cannot provide the original features and more background information about the data. Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-senstive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

### Tests of AutoEncoder -- Unsupervised Learning

```python
from tensorflow.keras.layers import Dense, Input, Activation, \
                                    Reshape, Flatten, Dropout, concatenate, BatchNormalization
from tensorflow.keras.models import Model, Sequential, load_model
from tensorflow.keras.utils  import to_categorical, plot_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics    import AUC
from tensorflow.keras.losses     import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras import regularizers

from sklearn.metrics import mean_squared_error, log_loss, r2_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix, \
                            roc_curve, auc, plot_roc_curve, \
                            precision_recall_curve, average_precision_score, plot_precision_recall_curve
                            
import tensorflow.keras.backend as K
```

```python
# FUNCTION
# Create an AutoEncoder model to reconstruct features

def create_autoencoder (input_shape, 
                        ecdr1_units, ecdr2_units, 
                        regularizer_rate=0.0, 
                        dropout_rate=0.0, 
                        learning_rate=0.01):
    
    # Encoders
    _inputs = Input(shape=input_shape, name="inputs")
    _flatten = Flatten(name="flatten")(_inputs)
    _encoder1 = Dense(units=ecdr1_units, activation="relu", 
                      kernel_initializer="random_uniform", 
                      kernel_regularizer=regularizers.l2(regularizer_rate), 
                      name="encoder1")(_flatten)
    _droput1  = Dropout(rate=dropout_rate, 
                        name="droput1")(_encoder1)
    _encoder2 = Dense(units=ecdr2_units, activation="selu", 
                      kernel_initializer="random_uniform", 
                      kernel_regularizer=regularizers.l2(regularizer_rate), 
                      name="encoder2")(_droput1)
    
    # Decoders
    dcdr2_units = ecdr1_units
    dcdr1_units = np.prod(input_shape)
    _decoder2 = Dense(units=dcdr2_units, activation="relu", 
                      kernel_initializer="random_uniform", 
                      kernel_regularizer=regularizers.l2(regularizer_rate), 
                      name="decoder2")(_encoder2)
    _decoder1 = Dense(units=dcdr1_units, activation="linear", 
                      kernel_initializer="random_uniform", 
                      kernel_regularizer=regularizers.l2(regularizer_rate), 
                      name="decoder1")(_decoder2)
    
    # Model
    model = Model(inputs=_inputs, outputs=_decoder1, 
                  name="autoencoder")
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, 
                  loss="mean_squared_error", 
                  metrics=["mean_squared_error"])
    
    return model
```

```python
input_shape = X_train.shape[1:]
ecdr1_units = 16
ecdr2_units = 8
regularizer_rate = 0.05
dropout_rate = 0.00  # NEVER Dropout!!!

autoencoder_mod = create_autoencoder(input_shape=input_shape, 
                                     ecdr1_units=ecdr1_units, ecdr2_units=ecdr2_units, 
                                     regularizer_rate=regularizer_rate, 
                                     dropout_rate=dropout_rate)
autoencoder_mod.summary()
```

```python
# FUNCTION
# Plot training scores

def plot_training_scores (hist, names_score):
    
    if isinstance(names_score, str):
        names_score = [names_score]
    num = len(names_score)
    
    plt.figure(figsize=(15,5))
    for i in range(num):
        plt.subplot(1, num, i+1)
        plt.plot(hist.history[names_score[i]], label=names_score[i])
        plt.xlabel("Epochs")
        plt.legend()
    plt.show()
    
    return
```

```python
# FUNCTION
# Mean Squared Error with Coefficients of parameters as Weights
#
# INPUT
# X_actu      : np.ndarray, 2D array (n_samples, n_parameters), actual values
# X_pred      : np.ndarray, 2D array (n_samples, n_parameters), predicted values
# coefs       : np.array, 1D array (n_parameters), weights of parameters
#
# OUTPUT
# mse_weights : np.array, 1D array (n_samples), weighted MSE of each sample

def mean_squared_error_weighted (X_actu, X_pred, coefs):
    
    X_actu = np.array(X_actu)
    X_pred = np.array(X_pred)
    coefs  = np.abs(np.array(coefs).reshape(-1))
    if ((X_actu.shape!=X_pred.shape) or (len(coefs)!=X_actu.shape[1])):
        print("Data Shape Invalid!")
        return
    
    n_samples    = X_actu.shape[0]
    n_parameters = X_actu.shape[1]
    mse_weighted = []
    
    for samp in range(n_samples):
        sse_samp = np.sum(np.power(np.multiply(coefs, np.subtract(X_actu[samp], X_pred[samp])), 2))
        mse_weighted.append(sse_samp / n_parameters)
    
    return np.array(mse_weighted)
```

```python
# FUNCTION
# Evaluation of Regression Prediction

def eval_regression (rgr_md, X_true, y_true, feature_weights=[], 
                     prt_flag=True):

    # Regression or Reconstruction
    is_reconstrunction = 0
    if (type(y_true)==type(X_true)):
        if ~(np.array(y_true)!=np.array(X_true)).sum():
            is_reconstrunction = 1
    
    # Prediction
    y_pred = rgr_md.predict(X_true)
    
    # is_reconstrunction ?
    if is_reconstrunction:
        if (feature_weights==[]):
            mse_y = mean_squared_error(np.transpose(y_true), np.transpose(y_pred),
                                       multioutput="raw_values")
        else:
            mse_y = mean_squared_error_weighted (y_true, y_pred, feature_weights)
        
        r2_y  = r2_score(np.transpose(y_true), np.transpose(y_pred),
                         multioutput="raw_values")
    else:
        mse_y = mean_squared_error(y_true, y_pred)
        r2_y  = r2_score(y_true, y_pred)

    if prt_flag:
        print("========== EVALUATION SCORES ==========\n")
        if is_reconstrunction:
            print(f'MSE.shape : [{mse_y.shape}]')
            print(f'R2 .shape : [{r2_y.shape}]')
        else:
            print(f'MSE : [{round(mse_y, 5)}]')
            print(f'R2  : [{round(r2_y,  5)}]')
    
    scores = dict({"MSE" : mse_y, 
                   "R2"  : r2_y})
    return (y_pred, scores)
```

```python
# FUNCTION
# Pick up observations
#
# INPUT
# obs_index  : 1D array, indexes of observations
# obs_value  : 1D array, values  of observations
# option     : 1 (by Numbers specified)
#              2 (by Percentage specified)
#              3 (by Threshold value specified)
# lower_tail : Boolean, lower or upper part to be pick
#
# OUTPUT
# rt         : dict, info of the pickup
# pck_index  : 1D array, indexes of pickup observations
# pck_value  : 1D array, values  of pickup observations

def select_observations (obs_index, obs_value, 
                         option=1, lower_tail=False, 
                         num_pck=100, pct_pck=0.01, val_pck=1.0):
    
    obs_index = np.array(obs_index)
    obs_value = np.array(obs_value)
    n_len = len(obs_index)
    if (len(obs_index) != len(obs_value)):
        print("ERROR: Input index and value do not match!")
        return
    
    dic_obs = dict(zip(obs_index, obs_value))
    obs_value.sort()
    
    # By Numbers
    if (option==1):
        if (lower_tail):
            val_pck = obs_value[num_pck-1]
        else:
            val_pck = obs_value[-num_pck]
        
    # By Percentage
    elif (option==2):
        num_pck = int(n_len * pct_pck)
        if (lower_tail):
            val_pck = obs_value[num_pck-1]
        else:
            val_pck = obs_value[-num_pck]

    # By Threshold Value
    elif (option==3):
        val_pck = val_pck
#         if (lower_tail):
#             num_pck = len(obs_value[obs_value <= val_pck])
#         else:
#             num_pck = len(obs_value[obs_value >= val_pck])
            
    else:
        print("ERROR: Option must be one of (1,2,3)!")
        return
    
    pck_index = []
    pck_value = []
    cnt_pck = 0
    
    if (lower_tail):
        for key, val in dic_obs.items():
            if (val <= val_pck):
                pck_index.append(key)
                pck_value.append(val)
                cnt_pck += 1
    else:
        for key, val in dic_obs.items():
            if (val >= val_pck):
                pck_index.append(key)
                pck_value.append(val)
                cnt_pck += 1
        
    rt = dict({"Number"     : cnt_pck,
               "Percentage" : round(cnt_pck/n_len, 5),
               "Threshold"  : val_pck,
               "lower_tail" : lower_tail})
    return (rt, np.array(pck_index), np.array(pck_value))
```

### Tests of Classification with Labeled Data -- Supervised Learning

```python
# FUNCTION
# Evaluation of Classification Prediction

def eval_classification (cls_md, X_true, y_true, 
                         prt_flag=True):
    
    y_pred_prob = cls_md.predict_proba(X_true)[:,1]  # <-- NOTICE HERE!
    y_pred = np.round(y_pred_prob, 0).astype(np.int)
    
    # Log-Loss
    logloss = -np.log(y_pred_prob)
    
    # Accuracy
    acr_score = accuracy_score(y_true, y_pred)
    if prt_flag:
        print("========== EVALUATION SCORES ==========\n")
        # Confusion Matrix
        print(classification_report(y_true, y_pred))

    # plot_roc_auc(cls_md, X_true, y_true)
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    auroc = auc(fpr, tpr)
    
    if prt_flag:
        plt.figure(figsize=(15,5))
        plt.subplot(1, 2, 1)
        plt.plot(fpr, tpr, label=f'AUC : {round(auroc, 3)}')
        plt.legend(loc='lower right')
        plt.plot((0,1), (0,1), 'g--')
        plt.title('ROC & AUC')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')

    # plot_precision_recall_curve(cls_md, X_true, y_true)
    precision, recall, thresholds = precision_recall_curve(y_true, y_pred_prob)
    auprc = average_precision_score(y_true, y_pred_prob)

    if prt_flag:
        plt.subplot(1, 2, 2)
        plt.plot(recall, precision, label=f'AUPRC : {round(auprc, 3)}')
        plt.legend(loc="lower left")
        plt.title("PRC & AUPRC")
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.show()

        print(f'SUM of Log-Loss: [{round(np.sum(logloss),0)}]')
        print(f'ACCURACY: [{round(acr_score, 3)}]')
        print(f'AUROC:    [{round(auroc, 3)}]')
        print(f'AUPRC:    [{round(auprc, 3)}]')

    scores = dict({"Accuracy" : acr_score, 
                   "AUROC"    : auroc, 
                   "AUPRC"    : auprc})
    
    return (y_pred, y_pred_prob, logloss, scores)
```

```python
# FUNCTION
# Remove Labeled Abnomalies from Resample

def resample_0 (X_train, y_train_labeled):

#    index_y_1 = y_train_labeled.index[y_train_labeled==1]
    index_y = y_train_labeled.index
    index_X_train = X_train.index
#     index_unlabels_bool = np.ones(len(index_X_train))
#     bool_unlabels = pd.Series(data=index_unlabels_bool, index=index_X_train)
#     for idx in index_1:
#         bool_unlabels[idx] = 0
#    index_unlabels = [i for i in index_X_train if i not in index_y_1]
    index_unlabels = [i for i in index_X_train if i not in index_y]
    X_train_unlabels = X_train.loc[index_unlabels]
    
    return X_train_unlabels
```

```python
# FUNCTION
# Re-Train AutoEncoder

def retrain_autoencoder (X_train):
    
    input_shape = X_train.shape[1:]
    ecdr1_units = int(input_shape[0] / 2)
    ecdr2_units = int(ecdr1_units / 2)
    regularizer_rate = 0.05
    dropout_rate = 0.0  # NEVER Dropout!!!
    autoencoder_mod = create_autoencoder(input_shape=input_shape, 
                                         ecdr1_units=ecdr1_units, ecdr2_units=ecdr2_units, 
                                         regularizer_rate=regularizer_rate, 
                                         dropout_rate=dropout_rate)

    n_epochs = 5
    batch_size = 1024
    hist = autoencoder_mod.fit(np.array(X_train), np.array(X_train), 
                               epochs=n_epochs, batch_size=batch_size, 
                               verbose=0)

    return (autoencoder_mod, hist)
```

```python
# FUNCTION
# Reconstruction of Features

def reconstruction_X (autoencoder_mod, X_train, feature_weights):
    
    # Keep Index for Observation Pick-up
    idx_X_train = X_train.index.values

    # Reconstruction of Features
    X_train_pred, scores = eval_regression(autoencoder_mod, 
                                           np.array(X_train), np.array(X_train), 
                                           feature_weights, 
                                           prt_flag=False)
    mse_X_train = scores["MSE"]

    # Loss (MSE) of Reconstruction
    X_recons_err = dict(zip(idx_X_train, mse_X_train))
    
    return (X_train_pred, X_recons_err)
```

```python
# FUNCTION
# Multi-round Reconstruction of Features

def reconstruction_X_repeat (X_train_unlabels, feature_weights, n_repeat=100):
    
    arr_X_train_pred = []
    arr_X_recons_err = []
    
    for n in range(n_repeat):
        # Train an AutoEncoder with X
        autoencoder_mod_rsp, hist = retrain_autoencoder (X_train_unlabels)

        # Reconstruction of the X
        X_train_pred, X_recons_err = reconstruction_X (autoencoder_mod_rsp, 
                                                       X_train_unlabels, feature_weights)
        
        arr_X_train_pred.append(X_train_pred)
        arr_X_recons_err.append(X_recons_err)
        
    return (arr_X_train_pred, arr_X_recons_err)
```

```python
# FUNCTION
# Pick Observations with High MSE of Feature Reconstruction

def pk_by_recons_err (X_recons_err, lower_tail=False,
                      option=1, num_pck=100, pct_pck=0.01, val_pck=2.5):

    # Pick-up by Number
    if (option==1):
        rt, pick_index, pick_value = select_observations(list(X_recons_err.keys()), 
                                                         list(X_recons_err.values()), 
                                                         lower_tail=lower_tail, 
                                                         option=option, num_pck=num_pck)
    # Pick-up by Percentage
    elif (option==2):
        rt, pick_index, pick_value = select_observations(list(X_recons_err.keys()), 
                                                         list(X_recons_err.values()), 
                                                         lower_tail=lower_tail, 
                                                         option=option, pct_pck=pct_pck)
    # Pick-up by Value-threshold
    elif (option==3):
        rt, pick_index, pick_value = select_observations(list(X_recons_err.keys()), 
                                                         list(X_recons_err.values()), 
                                                         lower_tail=lower_tail, 
                                                         option=option, val_pck=val_pck)

    # Makeup the Picked Observations
#     print("Pickup Observations:\n", rt)
    pick_mse_X_train = dict(zip(pick_index, pick_value))
    return (pick_mse_X_train)
```

```python
# FUNCTION
# Pick Observations with High MSE of Feature Reconstruction from Multi-rounds

def pk_by_recons_err_repeat (arr_X_recons_err, lower_tail=False, 
                             option=1, num_pck=100, pct_pck=0.01, val_pck=2.5):
    
    n_repeat = len(arr_X_recons_err)
    arr_pick_mse_X_train = []
    
    # Get an Array of Picked Observations
    for X_recons_err in arr_X_recons_err:
        pick_mse_X_train = pk_by_recons_err (X_recons_err, lower_tail=lower_tail,
                                             option=option, num_pck=num_pck)
        arr_pick_mse_X_train.append(pick_mse_X_train)
        
    # Get the Counts of Indexes of Observations
    dict_index_counts = {}
    for n in range(n_repeat):
        for idx, _ in arr_pick_mse_X_train[n].items():
            if (idx in dict_index_counts):
                dict_index_counts[idx] += 1
            else:
                dict_index_counts[idx] = 1
    
    # Sort the Indexes by Counts
    sort_index = sorted(dict_index_counts.items(), 
                        key=lambda x: x[1], reverse=True)
    
    # Return the top ranked Indexes
    index_pick_mse = []
    index_pick_mse_count = []

    pick_num = min(num_pck, len(sort_index))
    for i in range(pick_num):
        index_pick_mse.append(sort_index[i][0])
        index_pick_mse_count.append(sort_index[i][1])

### DEBUG
#     print((index_pick_mse[0], index_pick_mse_count[0]))
#     print((index_pick_mse[pick_num-1], index_pick_mse_count[pick_num-1]))
        
    return (index_pick_mse, index_pick_mse_count)
```

```python
# FUNCTION
# Merge and Filter Picked Observations from Reconstruction Errors and Prediction Loss

def pk_obs_merge_err_prob (index_pick_mse, index_pick_prob_1, index_pick_prob_05, 
                           num_pck, primary_mse=False, pct_prob_1=0.6):

    if (len(index_pick_prob_1)==0) or (len(index_pick_prob_05)==0):
        return (index_pick_mse[0:num_pck])
    
    index_2b_labeled = []
    index_pool = []

    # Priority is "index_pick_mse" 
    if (primary_mse):
        for i in range(num_pck):
            if (index_pick_mse[i] in index_pick_prob_1) or \
               (index_pick_mse[i] in index_pick_prob_05):
                index_2b_labeled.append(index_pick_mse[i])
            else:
                index_pool.append(index_pick_mse[i])
### DEBUG ###
        print(f'Overlapping: {len(index_2b_labeled)}')

        num_2b = len(index_2b_labeled)
        if (num_2b < num_pck):
            for j in range(num_pck-num_2b):
                index_2b_labeled.append(index_pool[j])
        return (index_2b_labeled)
    
    # Priority is "index_pick_prob_" 
    num_100 = int(num_pck * pct_prob_1)
    num_50  = num_pck - num_100

    # Pick Observations with Probabilities around 100%
    for i in range(num_100):
        if (index_pick_prob_1[i] in index_pick_mse):
            index_2b_labeled.append(index_pick_prob_1[i])
        else:
            index_pool.append(index_pick_prob_1[i])
### DEBUG ###
    print(f'Overlapping (100% & MSE): {len(index_2b_labeled)}')
            
    num_2b = len(index_2b_labeled)
    if (num_2b < num_100):
        for j in range(num_100-num_2b):
            index_2b_labeled.append(index_pool[j])
    index_2b_labeled_100 = index_2b_labeled.copy()
            
    # Pick Observations with Probabilities around 50%
    index_2b_labeled = []
    index_pool = []
    
    for i in range(num_50):
        if (index_pick_prob_05[i] in index_pick_mse):
            index_2b_labeled.append(index_pick_prob_05[i])
        else:
            index_pool.append(index_pick_prob_05[i])
### DEBUG ###
    print(f'Overlapping ( 50% & MSE): {len(index_2b_labeled)}')
            
    num_2b = len(index_2b_labeled)
    if (num_2b < num_50):
        for j in range(num_50-num_2b):
            index_2b_labeled.append(index_pool[j])
            
    return (index_2b_labeled_100 + index_2b_labeled)
```

```python
# FUNCTION
# Aggregate the Labeled datasets

def aggregate_labeled_data (X_train_labeled, y_train_labeled,
                            X_train_labeled_new, y_train_labeled_new):
    
    if (X_train_labeled is None) or (y_train_labeled is None):
        X_train_labeled = X_train_labeled_new
        y_train_labeled = y_train_labeled_new
    else:
        X_train_labeled = pd.concat([X_train_labeled, X_train_labeled_new], axis=0) 
        y_train_labeled = pd.concat([y_train_labeled, y_train_labeled_new], axis=0) 

    return (X_train_labeled, y_train_labeled)
```

```python
# FUNCTION
# Classification with the Labeled data;
# Then Predict Probabilities on Unlabeled data;
# Pickup Observations according to the Predicted Probabilities

def classification_prob (X_train_labeled, y_train_labeled, 
                         X_train_unlabels, num_pck=100):

    # Train a Classification model with the Labeled data
    Classifier_mod = XGBClassifier()
    Classifier_mod.fit(X_train_labeled, y_train_labeled)
    
    # Predict the Probabilities of Remained Unlabeled data
    resample_pred_prob = Classifier_mod.predict_proba(X_train_unlabels)

    # Get the observations with probabilities around 100%
    index_unlabels = X_train_unlabels.index
    resample_pred_prob_1 = resample_pred_prob[:, 1]
    rt, index_pick_prob_1, _ = select_observations(list(index_unlabels),
                                                   list(resample_pred_prob_1), 
                                                   lower_tail=False, 
                                                   num_pck=num_pck)

    # Get the observations with probabilities around 50%
    resample_pred_prob_diff = np.abs(np.subtract(resample_pred_prob[:, 0], 
                                                 resample_pred_prob[:, 1]))
#     resample_pred_loss = dict(zip(index_unlabels, resample_pred_prob_diff))
#     resample_pred_loss = sorted(resample_pred_loss.items(), 
#                                 key=lambda kv: (kv[1], kv[0]), reverse=False)
    rt, index_pick_prob_05_pool, _ = select_observations(list(index_unlabels),
                                                         list(resample_pred_prob_diff), 
                                                         lower_tail=True, 
                                                         num_pck= 2*num_pck)
    index_pick_prob_05 = []
    count_pck = 0
    for i in range(2*num_pck):
        if (index_pick_prob_05_pool[i] not in index_pick_prob_1):
            index_pick_prob_05.append(index_pick_prob_05_pool[i])
            count_pck += 1
            if (count_pck==num_pck):
                break
    return (list(index_pick_prob_1), list(index_pick_prob_05))
```

```python
# FUNCTION
# Get another Classification Model's Coefficients
# to Refine Reconstruction Error Calculation in the next round

def classification_coef (X_train_labeled, y_train_labeled):

    LR_mod = LogisticRegression()
    LR_mod.fit(X_train_labeled, y_train_labeled)

    return (LR_mod.coef_)
```

```python
# TRIAL_14

# This trial is an exhaustive approach
# 1) Starting from MSE (Unsupervised) as the primary Anomaly pickup base
# 2) Primary switching between MSE (Unsupervised) and Probabilities (Supervised) when either performs poorly
# 3) As the recognition Limit is unknown, the Iteration quits when having Switched too many (3) times

# primary_MSE = True (MSE as primary by default)
# primary_switch_count = 3 (Quit Iteration when Primary Switch the 3rd time)

# Iteraion Stop Criterion: pct_goal = 0.9


iteration_count = 0
total_y_label_1 = 0
count_y_label_1 = []
pct_goal = 0.9
num_pck  = 50
primary_switch_ratio = 0.1
primary_switch_alarm = False
primary_switch_count = 0
primary_MSE = True

X_train_unlabels = X_train.copy()
feature_weights = []
index_pick_prob_1  = []
index_pick_prob_05 = []
X_train_labeled = None
y_train_labeled = None


# Iteration
while total_y_label_1 < int(y_value_1_counts * train_size * pct_goal):

    # 1) Unsupervised Learning
    # Reconstruction of the Unlabeled data to get the Errors
    arr_X_train_pred, arr_X_recons_err = reconstruction_X_repeat (X_train_unlabels, 
                                                                  feature_weights,
                                                                  n_repeat=20)
    # Pickup Observations with HIGH Reconstruction Errors
    index_pick_mse,     _ = pk_by_recons_err_repeat (arr_X_recons_err, lower_tail=False, 
                                                     num_pck=num_pck)
    # Pickup Observations with LOW  Reconstruction Errors
    index_pick_mse_low, _ = pk_by_recons_err_repeat (arr_X_recons_err, lower_tail=True, 
                                                     num_pck=num_pck*(iteration_count+1))
    # Merge and Filter Picked Observations by Unsupervise Learning Errors
    # and Picked Observations by Supervised Learning Loss from the last round
    index_2b_labeled = pk_obs_merge_err_prob (index_pick_mse, index_pick_prob_1, index_pick_prob_05, 
                                              num_pck=num_pck, 
                                              primary_mse=primary_MSE, pct_prob_1=0.8)

    
    # 2) Human Labeling the Picked Observations
    X_train_labeled_new, y_train_labeled_new = human_labeling(X_train, y_train, 
                                                              index_2b_labeled)
    # "Fake Labeled 0" Observations
    X_train_fakelabel_0, y_train_fakelabel_0 = fake_labeling(X_train,  
                                                             index_pick_mse_low, fakelabel=0)
    # Aggregate the Labeled datasets
    X_train_labeled, y_train_labeled = aggregate_labeled_data (X_train_labeled, y_train_labeled, 
                                                               X_train_labeled_new, y_train_labeled_new)
    # Resampling from the whole dataset by removing the labeled data
    X_train_unlabels = resample_0 (X_train, y_train_labeled)
    
    
    # 3) Supervised Learning
    # Classification of the Labeled data ;
    # Then Predict "Loss" (Probabilities = 50%) on Unlabeled data;
    # Pickup Observations according to the Predicted Loss
    X_train_cls, y_train_cls = aggregate_labeled_data (X_train_labeled, y_train_labeled, 
                                                       X_train_fakelabel_0, y_train_fakelabel_0)
    index_pick_prob_1, index_pick_prob_05 = classification_prob (X_train_cls, y_train_cls, 
                                                                 X_train_unlabels, 
                                                                 num_pck=num_pck)

    
    # Iteration counts
    count_y_label_1.append(sum(pd.Series(y_train_labeled_new)==1))
    total_y_label_1 += count_y_label_1[iteration_count]
    count_y_label_1_ratio = count_y_label_1[iteration_count] / num_pck
    iteration_count += 1
### DEBUG    
    print (f'****** ROUND                     [{iteration_count}] ******')
    print (f'****** TOTAL LABELED DATA        [{len(X_train_labeled)}] ******')
    print (f'****** FAKE  LABELED DATA        [{len(X_train_fakelabel_0)}] ******')
    print (f'****** ROUND LABELED ABNORMALIES [{count_y_label_1[iteration_count-1]}] ******')
    print (f'****** TOTAL LABELED ABNORMALIES [{total_y_label_1}] ******\n')

    
    # 4) Supervised Learning_Additional
    # Get another Classification Model's Coefficients
    # to Refine Reconstruction Error Calculation in the next round
    if total_y_label_1:
        feature_weights = classification_coef (X_train_cls, y_train_cls)
    
    
    # 5) Primary Switch between MSE & Probabilities ?
    if (count_y_label_1_ratio < primary_switch_ratio):
        # Switch when two consecutive Pick-ups have Anomalies fewer than threshold
        if primary_switch_alarm:
            primary_MSE = not primary_MSE
            primary_switch_alarm = False
            primary_switch_count += 1
### DEBUG
            print("PRIMARY SWITCHED! MSE Primary(?):", primary_MSE)
            # Stop the Iteration when Switch more than twice
            if (primary_switch_count > 2):
                break
        # Alarm on when One Pick-up has Anomalies fewer than threshold
        else:
            primary_switch_alarm = True
    # Alarm off when One Pick-up has Anomalies more than threshold
    else:
        primary_switch_alarm = False

```

The plot demonstrates how the universal solution performs in the worst situation:
- The unknown Anomaly recognition Limit is reached after several Switches (vertical dash lines)
- The Iteration stops to avoid anymore unnecessary consumption of Human Labeling budget
- The approach unfortunately starts from MSE (Unsupervised Learning) as the primary for Anomaly selection

In common circumstances, the expected performance should fall into the range between this trial and the last trial, which fortunately starts from Probabilities (Supervised Learning) as the primary for Anomaly selection.
