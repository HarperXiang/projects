## Common Functions

### . Predictor Collinearity Cleaning

```python
# Function
# Remove collinear (alias) parameters
# 
# Input
# pDim   : integer > 1, dimension of matrix inputM
# inputM : 2-D np.array, square and symmetric matrix of predictors
# tol    : positive real, singularity tolerance
#
# Output
# A             : 2-D np.arrya, transferred matrix
# aliasParam    : alias parameters
# nonAliasParam : non-alias parameters

def SWEEPOperator(pDim, inputM, tol=0.1):  #tol=1e-10

    aliasParam    = []
    nonAliasParam = []
    A = np.array(inputM, copy=True, dtype=np.float)
    # diagA = np.diagonal(A)
    diagA = np.copy(np.diagonal(A))
    
    for k in range(pDim):
        akk = A[k,k]
        if (akk >= (tol * diagA[k])):
            nonAliasParam.append(k)
            for i in range(pDim):
                if (i != k):
                    for j in range(pDim):
                        if (j != k):
                            A[i,j] = A[i,j] - A[i,k] * (A[k,j] / akk)
                            A[j,i] = A[i,j]
                A[i,k] = A[i,k] / akk
                A[k,i] = A[i,k]
            A[k,k] = - 1.0 / akk
        else:
            aliasParam.append(k)
            for i in range(pDim):
                A[i,k] = 0.0
                A[k,i] = 0.0
    
    return A, aliasParam, nonAliasParam
```

### . Model Assessment

```python
# Function
# GLM model assessment metrics
#
# Inputs
# y  : pd.Series, observed  target values
# py : pd.Series, predicted target values
#
# Output
# outMetric: pd.Series that contains the metric values
#            RMSE   = Root Mean Squared Error
#            RELERR = Relative Error
#            RSQ    = R-Squared coefficient

def IntervalMetric(y, py):
    
    # Number of observations
    n = len(y)

    # Root Mean Squared Error
    ssq_error = np.sum((y - py) ** 2)
    rmse = np.sqrt(ssq_error / n)

    # Relative Error
    mean_y = np.mean(y)
    ssq_y  = np.sum((y - mean_y) ** 2)
    relerr = ssq_error / ssq_y

    # R-Squared
    mean_py = np.mean(py)
    ssq_py = np.sum((py - mean_py) ** 2)
    cssq_y_py = np.sum((y - mean_y) * (py - mean_py))
    rsq = (cssq_y_py**2) / (ssq_y * ssq_py)

    # Return
    outMetric = pd.Series({'RMSE'  : rmse,
                           'RELERR': relerr,
                           'RSQ'   : rsq})
    return (outMetric)
```

```python
# Function
# GLM model assessment Chi-Square Test
#
# Inputs
# y     : pd.Series, observed  target values
# py    : pd.Series, predicted target values
# distr : str, what Distribution, to choose Log-Likelihood function
# paras : Dictionary, parameters to calculate Log-Likelihood
#
# Output
# outChisq : pd.Series that contains the metric values
#            PEARSON  = Pearson Chi-Square
#            DEVIANCE = Deviance Chi-Square

def ModelChiSquare(y, py, distr, paras={}):

    # Pearson
    resid = y - py
    pearson = np.sum(resid**2 / py)

    # Deviance
    if (distr == "poisson"):
        dR2 = np.where(y_train > 0.0, 
                       y_train * np.log(y_train / y_pred), 0.0) - (y_train - y_pred)
        deviance = np.sum(2.0 * dR2)
    elif (distr == "gamma"):
        ratio = np.divide(y, py)
        GAMMA_Alpha = paras["GAMMA_Alpha"]
        dR2 = np.where(y > 0.0, 
                       2 * GAMMA_Alpha * (np.add(ratio, np.log(py)) - np.add(1, np.log(y))),
                       0.0)
        deviance = np.sum(dR2)

    # Return
    outChisq = pd.Series({'PEARSON' : pearson,
                          'DEVIANCE': deviance})
    return (outChisq)
```

```python
# Function
# Chi-Square Test of Deviance
#
# Inputs
# LLs : float, Log-Likelihood of models
# DFs : integer, degree of freedom of models
# list_distr_paras : List of Dictionaries, parameters of the Distribution
#
# Output
# Ds   : float, Deviance
# dDFs : float, difference of DFs
# PVs  : float, p-value of 1-cdf(D, df), upper tail probability of the Chi-squares distribution

from scipy.stats import chi2

def Chi2TestofDev(LLs, DFs, list_distr_paras):
    Ds   = []
    dDFs = []
    PVs  = []
    n = len(LLs)
    para_num = len(list_distr_paras[0].keys())

    # For Pure Premium model
    QLLKs = []
    PHIs  = [] 
    if (para_num and ("phi" in list_distr_paras[0].keys())):
        # First Unwrap list_distr_paras 
        for i in range(n):
            QLLKs.append(list_distr_paras[i]["qllk"])
            PHIs.append(list_distr_paras[i]["phi"])
        for i in range(1,n):
            D = (LLs[i] - LLs[0]) *2 / PHIs[0]
            df = int(-(DFs[i] - DFs[0]))
            Ds.append(D)
            dDFs.append(df)
            PVs.append(chi2.sf(D, df))
            
    # For Frequency and Severity models
    else:
        for i in range(1,n):
            D = (LLs[i] - LLs[0]) *2
            df = int(-(DFs[i] - DFs[0]))
            Ds.append(D)
            dDFs.append(df)
            PVs.append(chi2.sf(D, df))
    
    return (Ds, dDFs, PVs)
```

## Common Functions for Forward Selection

```python
# Function
# Plus each predictor separately in the list to the GLM model
# 
# Input
# GLM_model         : Object, Function of GLM model
# X0_name           : List, names of parameters in the existing model
# X0_measure        : List, types of parameters in the existing model
# X1_name           : List, names of candidate predictors
# X1_measure        : List, types of candidate predictors
# y_train           : np.array, y_train
# e_train           : np.array, Exposures as offset
# df_X              : pd.Dataframe, full dataset of X
# pre_paras         : Dictionary, pre-defined parameters of the distribution
#
# Output
# rt : pd.DataFrame incl. info
#      "Predictor"       : Predictors,
#      "nonAliasParaNum" : Number of Non-Alias Parameters,
#      "Log-Likelihood"  : Log-Likelihood values,
#      "DF"              : Degree of Freedoms
# list_distr_paras : List, multiple sets of parameters of the distribution

def PlusEachSinglePredictor (GLM_model, 
                             X0_name, X0_measure, X1_name, X1_measure, 
                             y_train, e_train, df_X, 
                             pre_paras={}):
    # Returns
    PARAMs = []
    Para_Nums = []
    LLs = []
    DFs = []
    list_distr_paras = []
    nObs = len(y_train)
    
    # GLM model with "_Intercept" only
    if (X0_name == []):
        iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras = GLM_model(None, y_train, e_train, 
                                                                                 pre_paras=pre_paras)
        outTable = pd.DataFrame(iterTable, columns = ['Iteration', 'Log-Likelihood', '_Intercept'])
        #print('Iteration History Table:\n', outTable)
        # Return values
        PARAMs.append('_Intercept')
        Para_Nums.append(len(nonAliasParam))
        LLs.append(outTable['Log-Likelihood'][len(outTable)-1])
        DFs.append(nObs - len(nonAliasParam))
        list_distr_paras.append(distr_paras)

    # When X0_name != NULL, get X0_train data
    for X_name, X_measure in zip(X0_name, X0_measure):
        # Handling Categorical Variables
        if (X_measure == 'categorical'):
            X_plusOne = pd.get_dummies(df_X[[X_name]].astype('category'))
        else:
            X_plusOne = df_X[[X_name]]
        # Combine all parameter in X0_name
        if (X_name == X0_name[0]):
            X0_train = X_plusOne
        else:
            X0_train = pd.concat([X0_train, X_plusOne], axis=1)
            
    # Run X0_train model
    if (X0_name != []):
        # GLM model ("_Intercept" added inside)
        iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras = GLM_model(X0_train, y_train, e_train,
                                                                                 pre_paras=pre_paras)
        # Return values
        PARAMs.append(['_Intercept']+X0_name)
        Para_Nums.append(len(nonAliasParam))
        LLs.append(iterTable[-1][1])
        DFs.append(nObs - len(nonAliasParam))
        list_distr_paras.append(distr_paras)
            
    # When X1_name != NULL, runing with each single Predictor
    for X_name, X_measure in zip(X1_name, X1_measure):
        # Handling Categorical Variables
        if (X_measure == 'categorical'):
            X_plusOne = pd.get_dummies(df_X[[X_name]].astype('category'))
        else:
            X_plusOne = df_X[[X_name]]
        # X0_train + One Predictor
        if (X0_name == []):
            X_train = X_plusOne
        else:
            X_train = pd.concat([X0_train, X_plusOne], axis=1)

        # GLM model ("_Intercept" added inside)
        iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras = GLM_model(X_train, y_train, e_train,
                                                                                 pre_paras=pre_paras)
        X_column = ['_Intercept'] + X_train.columns.tolist()
        param_name = [X_column[i] for i in nonAliasParam]
        outTable = pd.DataFrame(iterTable, columns = ['Iteration', 'Log-Likelihood'] + param_name)
        #print('Enter Predictors: ', X_name)
        #print('Parameters: ', param_name)
        #print('Number of Non-Aliased Parameters = ', len(param_name))
        #print('Iteration History Table:\n', outTable)
        #print('\n')
        
        # Return values
        PARAMs.append(X_name)
        Para_Nums.append(len(nonAliasParam))
        LLs.append(outTable['Log-Likelihood'][len(outTable)-1])
        DFs.append(nObs - len(nonAliasParam))
        list_distr_paras.append(distr_paras)
    
    # Return 
    rt = pd.DataFrame({"Predictor"      : PARAMs,
                       "nonAliasParaNum": Para_Nums,
                       "Log-Likelihood" : LLs,
                       "DF"             : DFs})
    return (rt, list_distr_paras)
```

```python
# Function
# Auto Forward Selection step by step and make summary
#
# Input
# GLM_mode      : Object, Function of GLM model
# X0_name       : List, names of parameters in the existing model
# X0_measure    : List, types of parameters in the existing model
# X1_name       : List, names of candidate predictors
# X1_measure    : List, types of candidate predictors
# y_train       : np.array, y_train
# e_train       : np.array, Exposures as offset
# summary_fwd   : pd.DataFrame, summary of _Intercept Only
# df_X          : pd.Dataframe, full dataset of X
#
# Output
# summary_fwd   : pd.DataFrame, summary

def ForwardSelection (GLM_mode, 
                      X0_name, X0_measure, X1_name, X1_measure, y_train, e_train, 
                      summary_fwd, df_X,
                      pre_paras={}):

    ct_round = 0
    while (len(X1_name)):
        ct_round += 1

        # Checking each one Predictor added to the Existing Parameters
        rt, list_distr_paras = PlusEachSinglePredictor (GLM_mode, 
                                                        X0_name, X0_measure, X1_name, X1_measure, 
                                                        y_train, e_train, df_X=df_X, 
                                                        pre_paras=pre_paras)
        Ds, dDFs, PVs = Chi2TestofDev(rt["Log-Likelihood"], rt["DF"], list_distr_paras)
        rt["Deviance"] = [np.nan] + Ds
        rt["Diff_DF"]  = [np.nan] + dDFs
        rt["P-value"]  = [np.nan] + PVs
        print(f'Round {ct_round}:')
        print(rt)
        print ("\n")

        # Get the Parameter with the minimum P-value
        para_minP = rt[rt["P-value"]==rt["P-value"].min()]
        if (len(para_minP)>1):
            para_minP = para_minP.iloc[[0]]

        # Summary
        summary_fwd = pd.concat([summary_fwd, para_minP], axis=0)

        # Updating Parameters for the next Round
        para_added = summary_fwd["Predictor"].tolist()[-1]
        X0_name    = X0_name + [para_added]
        X0_measure = X0_measure + [X1_measure[X1_name.index(para_added)]]
        X1_measure.pop(X1_name.index(para_added))
        X1_name.remove(para_added)
        
    # Return
    return (summary_fwd)
```

# Frequency Modeling

A Frequency model to predict the number of claims with the following specifications:

1.	The target variable is num_claims

2.	The predictors are f_primary_age_tier, f_primary_gender, f_marital, f_residence_location, f_fire_alarm_type, f_mile_fire_station, and f_aoi_tier

3.	The distribution assumption is Poisson

4.	The link function is the logarithm

5.	The offset variable is the logarithm of exposure

6.	The model will include the Intercept term

7.	Use the Forward Selection method to only enter significant predictors

8.	Use the Deviance statistic to select the predictor entered in each step

9.	The tolerance level is 5%

### . Manual Modeling Function

```python
# Function
# General Linear Model for Frequecy
#
# Input
# X_train   : pd.Dataframe, X_train, predictors
# y_train   : pd.Series,    y_train, target varialbe (Claims)
# exposure  : pd.Series,    Exposures
# pre_paras : Dictionary,   pre-defined parameters of the Distribution (NOT used in this function)
#
# Output
# iterTable   : 2D List, Iteration info
# beta        : np.array, Coefficients of predictors, 
# covBeta     : 2D np.array, Covariance of beta
# y_pred      : np.array, predicted target variable
# distr_paras : Dictionary, output parameters of the Distribution

def GLM4Frequency(X_train, y_train, exposure, pre_paras={},
                  maxIter=20, maxStep=5, tolLLK=1e-3, tolBeta=1e-10):
    
    nObs = len(y_train)

    # Get Offset: ln(T)
    offset = np.log(exposure.values)

    # Precompute the ln(y!)
    constLLK = scipy.special.loggamma(y_train+1.0)

    # Generate the design matrix with the Intercept term as the first column
    if X_train is not None:
        Xp1 = X_train.copy()
        Xp1.insert(0, '_Intercept', 1.0)
    else:
        Xp1 = pd.DataFrame(np.full((nObs,1), 1.0), columns = ['_Intercept'] )
    
    # Make all objects as np
    designX  = Xp1.values
    designXT = np.transpose(designX)

    # Find the non-aliased columns
    pDim = Xp1.shape[1]
    inputM = np.dot(designXT, designX)
    outputM, aliasParam, nonAliasParam = SWEEPOperator (pDim, inputM)
    designX  = Xp1.values[:, nonAliasParam]
    designXT = np.transpose(designX)

    # Initialize Log-Likelihood
    beta = np.zeros((len(nonAliasParam)))
    beta[0] = np.log(np.mean(y_train))
    nu = offset + designX.dot(beta)
    y_pred = np.exp(nu)
    llk = np.sum(y_train * nu - y_pred - constLLK)

    # Prepare the iteration history table
    itList = [0, llk]
    for b in beta:
       itList.append(b)
    iterTable = [itList]

    # Iteration
    for it in range(maxIter):
        
        # Fisher's Information
        gradient =   np.dot(designXT, (y_train - y_pred))
        # Here using reshape() instead of diagonal matrix for y_pred to save computation space
        hessian  = - np.dot(designXT, (y_pred.reshape((nObs,1)) * designX))
        
        # Taylor Expansion
        # delta = _n+1  _n =  u(n) / I(n)
        delta = - np.linalg.solve(hessian, gradient)
        
        step = 1.0
        for iStep in range(maxStep):
            # _n+1
            beta_next = beta + step * delta
            #  = Xt  + offset
            nu_next = offset + np.dot(designX, beta_next)
            # Inverse Cannonical Link
            y_pred_next = np.exp(nu_next)
            # Log-Likelihood
            llk_next = np.sum(y_train * nu_next - y_pred_next - constLLK)
            
            # Practically guarantee LLK is increasing (at least not decreasing too much due to errors)
            if ((llk_next - llk) > - tolLLK):
                break
            # In case of decreasing too much, try again with a smaller delta
            else:
                step = 0.5 * step
                
        # For the next round of Iteration
        diffBeta = beta_next - beta
        llk = llk_next
        beta = beta_next
        y_pred = y_pred_next
        
        # Collect the Iteration info
        itList = [it+1, llk]
        for b in beta:
            itList.append(b)
        iterTable.append(itList)
        
        # Convergency of _n+1 and _n
        if (np.linalg.norm(diffBeta) < tolBeta):
            break

    # Final Covariance matrix
    hessian = - np.dot(designXT, (y_pred.reshape((nObs,1)) * designX))
    covBeta = - np.linalg.inv(hessian)

    # Return
    distr_paras = {}
    return(iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras)
```

### . Manual Modeling Function

```python
# Function
# Solve this equation for Alpha
#
# Input
# c      : positive real number, value of equation
#          log(alpha) - digamma(alpha) = c
#
# Output
# alpha  : positive real number, solved value

def solve4Alpha (c, maxIter = 100, epsilon = 1e-10):

    # Find a0 such that f0 is greater than or equal to c
    a0 = 0.5
    while True:
        f0 = np.log(a0) - scipy.special.digamma(a0)
        if (f0 < c):
            a0 = a0 / 2.0
        else:
            break
 
    # Find a1 such that f1 is less than or equal to c
    a1 = 2.0
    while True:
        f1 = np.log(a1) - scipy.special.digamma(a1)
        if (f1 > c):
            a1 = a1 * 2.0
        else:
            break

    # Update the end-points
    for nIter in range(maxIter):
        alpha = (a0 + a1) / 2.0
        func = np.log(alpha) - scipy.special.digamma(alpha)
        if (abs(func-c) > epsilon):
            if (func > c):
                a0 = alpha
            else:
                a1 = alpha
        else:
            break

    return (alpha)
```

```python
# Function
# General Linear Model for Severity
#
# Input
# X_train   : pd.Dataframe, X_train, predictors
# y_train   : pd.Series,    y_train, target varialbe (Severity)
# exposure  : pd.Series,    Exposures
# pre_paras : Dictionary,   pre-defined parameters of the Distribution (NOT used in this function)
#
# Output
# iterTable   : 2D List, Iteration info
# beta        : np.array, Coefficients of predictors, 
# covBeta     : 2D np.array, Covariance of beta
# y_pred      : np.array, predicted target variable
# distr_paras : Dictionary, output parameters of the Distribution

def GLM4Severity(X_train, y_train, exposure, pre_paras={},
                 maxIter=20, maxStep=5, tolLLK=1e-3, tolBeta=1e-10):
    
    nObs = len(y_train)

    # Get Offset: ln(T)
    #offset = np.log(exposure.values)

    # Generate the design matrix with the Intercept term as the first column
    if X_train is not None:
        Xp1 = X_train.copy()
        Xp1.insert(0, '_Intercept', 1.0)
    else:
        Xp1 = pd.DataFrame(np.full((nObs,1), 1.0), columns = ['_Intercept'] )
    
    # Make all objects as np
    designX  = Xp1.values
    designXT = np.transpose(designX)
    y_train  = y_train.values

    # Find the non-aliased columns
    pDim = Xp1.shape[1]
    inputM = np.dot(designXT, designX)
    outputM, aliasParam, nonAliasParam = SWEEPOperator (pDim, inputM)
    designX  = Xp1.values[:, nonAliasParam]
    designXT = np.transpose(designX)

    # Initialize Log-Likelihood
    beta = np.zeros((len(nonAliasParam)))
    beta[0] = np.log(np.mean(y_train))
    nu = designX.dot(beta)
    y_pred = np.exp(nu)
    ratio = np.divide(y_train, y_pred)

    # Here's the profile (simplified) Log-Likelihood with beta only
    #llk = - np.sum(ratio + np.log(y_pred))
    # Here's the Log-Likelihood with beta and alpha
    c = np.mean(np.subtract(ratio, np.log(ratio))) - 1.0
    alpha = solve4Alpha (c)
    constLLK = scipy.special.loggamma(alpha)    # ln(())
    llk = np.sum(alpha * -np.add(ratio, np.log(y_pred)) +
                 (alpha - 1.0) * np.log(y_train) +
                 alpha * np.log(alpha) -
                 constLLK)
    
    # Prepare the iteration history table
    itList = [0, llk]
    for b in beta:
       itList.append(b)
    iterTable = [itList]

    # Iteration
    for it in range(maxIter):
        
        # Fisher's Information
        gradient =   np.dot(designXT, np.subtract(ratio, 1))
        # Here using reshape() instead of diagonal matrix for y_pred to save computation space
        hessian  = - np.dot(designXT, (ratio.reshape((nObs,1)) * designX))
        
        # Taylor Expansion
        # delta = _n+1  _n =  u(n) / I(n)
        delta = - np.linalg.solve(hessian, gradient)
        
        step = 1.0
        for iStep in range(maxStep):
            # _n+1
            beta_next = beta + step * delta
            #  = Xt 
            nu_next = np.dot(designX, beta_next)
            # Inverse Log Link
            y_pred_next = np.exp(nu_next)
            # Ratio of y/y.hat
            ratio_next = np.divide(y_train, y_pred_next)
            
            # profile Log-Likelihood
            #llk_next = - np.sum(ratio_next + np.log(y_pred_next))
            # Here's the Log-Likelihood with beta and alpha
            c = np.mean(np.subtract(ratio_next, np.log(ratio_next))) - 1.0
            alpha = solve4Alpha (c)
            constLLK = scipy.special.loggamma(alpha)    # ln(())
            llk_next = np.sum(alpha * -np.add(ratio_next, np.log(y_pred_next)) +
                              (alpha - 1.0) * np.log(y_train) +
                              alpha * np.log(alpha) -
                              constLLK)
            
            # Practically guarantee LLK is increasing (at least not decreasing too much due to errors)
            if ((llk_next - llk) > - tolLLK):
                break
            # In case of decreasing too much, try again with a smaller delta
            else:
                step = 0.5 * step
                
        # For the next round of Iteration
        diffBeta = beta_next - beta
        llk    = llk_next
        beta   = beta_next
        y_pred = y_pred_next
        ratio  = ratio_next
        
        # Collect the Iteration info
        itList = [it+1, llk]
        for b in beta:
            itList.append(b)
        iterTable.append(itList)
        
        # Convergency of _n+1 and _n
        if (np.linalg.norm(diffBeta) < tolBeta):
            break

    # Final Covariance matrix
    hessian  = - np.dot(designXT, (ratio.reshape((nObs,1)) * designX))
    covBeta  = - np.linalg.inv(hessian) / alpha

    # Return
    distr_paras = {"GAMMA_Alpha": alpha}
    return(iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras)
```

# Pure Premium Prediction Solution_2: Direct Modeling

A Pure Premium model to predict the total amount of claims with the following specifications:

1.	The target variable is amt_claims

2.	The predictors are f_primary_age_tier, f_primary_gender, f_marital, f_residence_location, f_fire_alarm_type, f_mile_fire_station, and f_aoi_tier

3.	The distribution assumption is Tweedie with your p parameter in Question 5

4.	The link function is the logarithm

5.	The offset variable is the logarithm of exposure

6.	The model will include the Intercept term

7.	Use the Forward Selection method to only enter significant predictors

8.	Use the Deviance statistic to select the predictor entered in each step

9.	The tolerance level is 5%

### . Manual Modeling Functions

```python
# Function
# General Linear Model for Pure Premium
#
# Input
# X_train   : pd.Dataframe, X_train, predictors
# y_train   : pd.Series,    y_train, target varialbe (Loss)
# exposure  : pd.Series,    Exposures
# pre_paras : Dictionary,   pre-defined parameters of the Distribution
#
# Output
# iterTable   : 2D List, Iteration info
# beta        : np.array, Coefficients of predictors, 
# covBeta     : 2D np.array, Covariance of beta
# y_pred      : np.array, predicted target variable
# distr_paras : Dictionary, output parameters of the Distribution

def GLM4PurePremium (X_train, y_train, exposure, pre_paras={"tweedieP":1.5}, 
                            maxIter=50, maxStep=5, tolLLK=1e-3, tolBeta=1e-10):

    tweedieP = pre_paras["tweedieP"]
    nObs = len(y_train)
    #yvec = y_train.to_numpy(dtype = float)
    yvec = np.array(y_train)
    #ovec = np.log(exposure.to_np(dtype = float))
    ovec = np.log(np.array(exposure))
    two_p = 2.0 - tweedieP
    one_p = 1.0 - tweedieP
    ypow21 = np.power(y_train, two_p) / two_p / one_p 

    # Generate the design matrix with the Intercept term as the first column
    if X_train is not None:
        Xp1 = X_train.copy()
        Xp1.insert(0, '_Intercept', 1.0)
    else:
        Xp1 = pd.DataFrame(np.full((nObs,1), 1.0), columns = ['_Intercept'] )

    pDim = Xp1.shape[1]
    designX = Xp1.values
    designXT = np.transpose(designX)
    
    # Find the non-aliased columns
    inputM = np.dot(designXT, designX)
    outputM, aliasParam, nonAliasParam = SWEEPOperator (pDim, inputM)

    # Make all objects as np
    designX = Xp1.values[:,nonAliasParam]
    designXT = np.transpose(designX)

    # Initialize beta array
    beta = np.zeros((len(nonAliasParam)))
    beta[0] = np.log(np.mean(yvec))
    nu = ovec + designX.dot(beta)
    y_pred = np.exp(nu)
    powvec = np.power(y_pred, one_p)
    devvec = (ypow21 - yvec * powvec / one_p + y_pred * powvec / two_p)
    qllk = - np.sum(devvec)

    # Prepare the iteration history table
    itList = [0, qllk]
    for b in beta:
        itList.append(b)
    iterTable = [itList]

    for it in range(maxIter):
        rvec = (y_pred - yvec) * powvec
        gradient = 2.0 * np.dot(designXT, rvec)
        svec = (two_p * y_pred - one_p * yvec) * powvec
        hessian = 2.0 * np.dot(designXT, (svec.reshape((nObs,1)) * designX))
        delta = np.linalg.solve(hessian, gradient)
        step = 1.0
        for iStep in range(maxStep):
            beta_next = beta - step * delta
            nu_next = ovec + np.dot(designX, beta_next)
            y_pred_next = np.exp(nu_next)
            powvec_next = np.power(y_pred_next, one_p)
            devvec = (ypow21 - yvec * powvec_next / one_p + y_pred_next * powvec_next / two_p)
            qllk_next = - np.sum(devvec) 
            if ((qllk_next - qllk) > - tolLLK):
                break
            else:
                step = 0.5 * step
        diffBeta = beta_next - beta
        qllk = qllk_next
        beta = beta_next
        powvec = powvec_next
        y_pred = y_pred_next
        itList = [it+1, qllk]
        for b in beta:
            itList.append(b)
        iterTable.append(itList)
        if (np.linalg.norm(diffBeta) < tolBeta):
            break

    svec = ((two_p * y_pred - one_p * yvec) * powvec)
    hessian = 2.0 * np.dot(designXT, (svec.reshape((nObs,1)) * designX))

    devvec = (ypow21 - yvec * powvec / one_p + y_pred * powvec / two_p)
    phi = np.sum(devvec) / (nObs - len(nonAliasParam))
    covBeta = np.linalg.inv(hessian) / phi
    distr_paras = {"qllk": qllk,
                   "phi" : phi}

#    return(iterTable, nonAliasParam, qllk, phi, beta, covBeta, y_pred)
    return(iterTable, nonAliasParam, beta, covBeta, y_pred, distr_paras)
```

```python
# Final parameter estimates
outCoefficient = pd.Series(beta, index = param_name)
print('Parameter Estimates:\n', outCoefficient)
```

```python
# Final covariance matrix
outCovb = pd.DataFrame(covBeta, index = param_name, columns = param_name)
outCovb
```

```python
# Standard Error of parameters
se_coeff = []
for i in range(len(param_name)):
    se_coeff.append(np.sqrt(outCovb.iloc[i,i]))

SE_Coefficient = pd.Series(se_coeff, index = param_name)
SE_Coefficient
```

```python
# Corrlation matrix
outCorr = outCovb.copy()

for i in range(len(param_name)):
    for j in range(len(param_name)):
        outCorr.iloc[i,j] = outCovb.iloc[i,j] / (SE_Coefficient[i]*SE_Coefficient[j])
outCorr
```

```python
# Confidence Interval of parameters
alpha = 0.05
z = scipy.stats.norm.ppf(1-alpha/2)
CI_Low = []
CI_Upp = []

for i in range(len(param_name)):
    L = outCoefficient[i] - z*SE_Coefficient[i]
    U = outCoefficient[i] + z*SE_Coefficient[i]
    CI_Low.append(L)
    CI_Upp.append(U)

smy_Coefficient = pd.DataFrame({"Estimate": outCoefficient,
                               "Standard_Error": SE_Coefficient,
                               "CI_Lower": CI_Low,
                               "CI_Upper": CI_Upp}, 
                               index=param_name)
smy_Coefficient
```

```python
# All Parameters incl. Non-alias and Aliased
param_full = pd.DataFrame({"Parameter"          : X_column,
                           "EstimateValue"      : np.zeros(len(X_column)),
                           "exp(EstimateValue)" : np.ones(len(X_column))})

parameters = outTable.iloc[len(outTable)-1][1:]
predictors = parameters.index.tolist()

for prdt in predictors:
    param_full.loc[param_full["Parameter"]==prdt, "EstimateValue"] = parameters[prdt]
param_full["exp(EstimateValue)"] = np.exp(param_full["EstimateValue"])

# Add the Phi parameter of Tweedie distribution
param_full.loc[len(param_full)] = ["Tweedie_Phi", distr_paras["phi"], np.nan]
param_full
```

# Model Comparison (Two-way Lift Chart)

Compare these two (indirect vs. direct) Pure Premium predictions models

1.	The Poisson Frequency and Gamma Severity model

2.	The Tweedie Pure Premium model

The Impact is calculated by dividing the scaled Pure Premium of Model 2 by that of Model 1.  Please present the table of values that you use to generate the chart.  The group indices are on the row, the columns contain the minimum and the maximum impact values, the total observed and predicted amounts of claims, and the models loss ratios.

### . Verification

```python
LS0 = inputData[name_Loss]
LS1 = np.multiply(predPurePremium_S1, np.array(e_train))
LS2 = np.multiply(predPurePremium_S2, np.array(e_train))

LS_sum_table = pd.DataFrame({name_Exposure        : e_train.sum(),
                             name_Loss            : sum(LS0),
                             "predLoss_S1=Premium*exposure" : sum(LS1),
                             "predLoss_S2=Premium*exposure" : sum(LS2),
                             "LossRatio_S1"       : sum(LS0)/sum(LS1),
                             "LossRatio_S2"       : sum(LS0)/sum(LS2)},
                           index=["TOTAL"])
LS_sum_table
```

### Rescale Predicted Loss

```python
baseline_LS1 = sum(LS0) / sum(LS1)
re_LS1 = baseline_LS1 * LS1

baseline_LS2 = sum(LS0) / sum(LS2)
re_LS2 = baseline_LS2 * LS2

print(baseline_LS1, sum(re_LS1), baseline_LS2, sum(re_LS2))
```

```python
Impact_table = pd.DataFrame({name_Exposure : e_train,
                             name_Loss     : inputData[name_Loss],
                             "predLoss_S1" : re_LS1,
                             "predLoss_S2" : re_LS2})
Impact_table.shape
```

```python
Impact_table["Impact"] = Impact_table["predLoss_S2"] / Impact_table["predLoss_S1"]
Impact_table = Impact_table.sort_values(["Impact"], ascending=True)
Impact_table
```

```python
# Divide into 10 groups with equal SUM of Exposures
gp_limExposure = Impact_table[name_Exposure].sum() / 10

gp_sumExposure = 0
gp_nums  = []
gp_count = 0

for i in range(len(Impact_table)):
    gp_sumExposure += np.float(Impact_table.iloc[i][name_Exposure])
    gp_count += 1
    if (gp_sumExposure >= gp_limExposure):
        gp_nums.append(gp_count)
        gp_sumExposure = 0
        gp_count = 0
    if (i==len(Impact_table)-1 and gp_count):
        gp_nums.append(gp_count)

print(len(gp_nums), sum(gp_nums))
```

```python
# Gather data of each Group
gp_start = 0
gp_Exposure     = []
gp_obsLoss      = []
gp_minImpact    = []
gp_maxImpact    = []
gp_predLoss_S1  = []
gp_predLoss_S2  = []
gp_LossRatio_S1 = []
gp_LossRatio_S2 = []

for n in gp_nums:
    # Each Group
    gp = Impact_table.iloc[gp_start : (gp_start+n)]
    #print(gp_start, n, len(gp))
    
    # Sum of Exposures
    gp_Exposure.append(gp[name_Exposure].sum())
    
    # Sum of Observed Loss
    obsLoss = gp[name_Loss].sum()
    gp_obsLoss.append(obsLoss)
    
    # Impact range
    gp_minImpact.append(gp.iloc[0]["Impact"])
    gp_maxImpact.append(gp.iloc[n-1]["Impact"])
    
    # Sum of Predicted Loss
    predLoss_S1 = gp["predLoss_S1"].sum()
    predLoss_S2 = gp["predLoss_S2"].sum()
    gp_predLoss_S1.append(predLoss_S1)
    gp_predLoss_S2.append(predLoss_S2)
    
    # Loss Ratio 
    gp_LossRatio_S1.append(obsLoss / predLoss_S1)
    gp_LossRatio_S2.append(obsLoss / predLoss_S2)
    
    # Next Group starts
    gp_start += n

# Create Lift Chart
LC_2way = pd.DataFrame({"Num_Observations" : gp_nums,
                        name_Exposure      : gp_Exposure,
                        name_Loss          : gp_obsLoss,
                        "Impact_min"       : gp_minImpact,
                        "Impact_max"       : gp_maxImpact,
                        "predLoss_S1"      : gp_predLoss_S1,
                        "predLoss_S2"      : gp_predLoss_S2,
                        "Loss_Ratio_S1"    : gp_LossRatio_S1,
                        "Loss_Ratio_S2"    : gp_LossRatio_S2})
LC_2way = LC_2way.set_index(LC_2way.index+1)
LC_2way
```

```python
# Verification
print(LC_2way[name_Exposure].sum(), Impact_table[name_Exposure].sum())
print(LC_2way[name_Loss].sum(), Impact_table[name_Loss].sum())
print(LC_2way["predLoss_S1"].sum(), Impact_table["predLoss_S1"].sum())
print(LC_2way["predLoss_S2"].sum(), Impact_table["predLoss_S2"].sum())
```

```python
# Distance to 1
LC_2way["Distance_S1"] = LC_2way["Loss_Ratio_S1"] - 1
LC_2way["Distance_S2"] = LC_2way["Loss_Ratio_S2"] - 1
LC_2way
```

```python
dist_S1 = sum(LC_2way["Distance_S1"].abs())
dist_S2 = sum(LC_2way["Distance_S2"].abs())

print("Distance of Solution_1:", dist_S1)
print("Distance of Solution_2:", dist_S2)
if dist_S1 < dist_S2:
    print("\n=== WINNER is Solution_1, F*S Model ===")
else:
    print("\n=== WINNER is Solution_2: Tweedie Model ===")
```
